## Problem Statement:
Perform exploratory analysis on e-commerce customer behavior to uncover patterns in browsing time, cart activity, and purchase likelihood.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"E:\PGDA 41\Machine Learning\assignment 2\E-Commerce Customer Behavior Analysis\ecommerce_customer_behavior.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
sns.countplot(data=df, x="PurchaseMade")
plt.title("Purchase Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="PurchaseMade", y="BrowsingTime", data=df)
plt.title("Browsing Time vs Purchase")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="CartItems", y="PurchaseMade", data=df, estimator=lambda x: sum(x)/len(x))
plt.title("Average Purchase Rate by Cart Items")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=["float64", "int64"]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df, x="PageViews", hue="PurchaseMade", kde=True, bins=20, multiple="stack")
plt.title("Page Views vs Purchase")
plt.show()
```
