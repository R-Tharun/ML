## Problem Statement:
Predict whether a passenger survived or not based on demographic and travel-related features such as class, age, sex, and embarkation port.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"E:\PGDA 41\Machine Learning\assignment 2\Cereal Data  Analysis\cereal_data_clean.csv")
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Rating'], kde=True)
plt.title("Cereal Rating Distribution")
```

```python
numeric_df = df.select_dtypes(include='number')
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
brand_avg = df.groupby("Brand")["Rating"].mean().sort_values(ascending=False)
print(brand_avg)
```

```python
healthy = df[(df["Fiber"] > 5) & (df["Sugars"] < 5)]
print(healthy[["Cereal", "Fiber", "Sugars", "Rating"]])
```
